```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import glob
import os
```

```python
files = glob.glob("snow_data/*.csv")

# Initialize an empty list to store DataFrames
df_list = []

for file in files:
    df = pd.read_csv(file)
    mountain_name = os.path.basename(file).split('.')[0]
    df['Resort'] = mountain_name
    df_list.append(df)

final_df = pd.concat(df_list, ignore_index=True)
final_df
```

```python
# Assuming final_df is the combined dataframe
# Step 1: Clean the 'Year' column to extract just the first year
final_df['Year'] = final_df['Year'].apply(lambda x: str(x).split(' - ')[0])

final_df['Year'] = final_df['Year'].astype(int)
# Step 2: Clean 'Total Snowfall', 'Average Base Depth', 'Max Base Depth', 'Biggest Snowfall' columns by removing quotes and converting to integers
final_df['Total Snowfall'] = final_df['Total Snowfall'].replace('"', '', regex=True).astype(float)
final_df['Snowfall Days'] = final_df['Snowfall Days'].replace(' days', '', regex=True).astype(int)
final_df['Average Base Depth'] = final_df['Average Base Depth'].replace('"', '', regex=True).astype(float)  # Keep as float for possible decimals
final_df['Max Base Depth'] = final_df['Max Base Depth'].replace('"', '', regex=True).astype(float)
final_df['Biggest Snowfall'] = final_df['Biggest Snowfall'].replace('"', '', regex=True).astype(float)

# Step 3: Rearrange the columns to match the desired order
final_df = final_df[['Year', 'Total Snowfall', 'Snowfall Days', 'Average Base Depth', 'Max Base Depth', 'Biggest Snowfall', 'Resort']]

# Step 4: Optional - Save the cleaned data to a new CSV file
# final_df.to_csv("cleaned_mountain_data.csv", index=False)

# Filter for the years between 2014 and 2024 (inclusive)
final_df = final_df[(final_df['Year'] >= 2014) & (final_df['Year'] <= 2024)]

# Show the filtered DataFrame
final_df
```

```python
snow_name = final_df.Resort.unique().tolist()
```

```python
df = pd.read_csv('climate_visitor.csv')
df.Resort.unique()
df_name = df.Resort.unique().tolist()
name_mapping = {
    'charlotte_pass': 'Charlotte Pass',
    'fall_creek': 'Falls Creek',
    'mt_baw_baw': 'Mt. Baw Baw',
    'mt_buller': 'Mt. Buller',
    'mt_hotham': 'Mt. Hotham',
    'mt_selwyn': 'Selwyn',
    'perish': 'Perisher',
    'thredbo': 'Thredbo'
}
final_df['Resort'] = final_df['Resort'].map(name_mapping)
final_df
```
